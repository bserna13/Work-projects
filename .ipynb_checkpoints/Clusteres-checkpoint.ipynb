## Dendrograma 

```python
# Creamos el dendograma para encontrar el número óptimo de clusters

import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(data_pro, method = 'ward'))

plt.title('Dendograma')
plt.xlabel('Clientes')
plt.ylabel('Distancias Euclidianas')
plt.show()
```

#### Vamos a hallar el valor de K haciendo una gráfica e intentando hallar el “punto de codo” que comentábamos antes. Este es nuestro resultado:

```python
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min


Nc = range(1, 20)
kmeans = [KMeans(n_clusters=i) for i in Nc]
kmeans
score = [kmeans[i].fit(data_pro).score(data_pro) for i in range(len(kmeans))]
score
plt.plot(Nc,score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
pd.DataFrame({"Nc":Nc,"Score":score})
```

### El numero de observaciones que pareciera optimo es 5

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
import matplotlib.cm as cm


# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data_pro)

# Realizar PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Realizar clustering (reemplaza n_clusters con el número de clusters deseado)
kmeans = KMeans(n_clusters=4, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

```python
# Obtener los colores únicos para cada cluster
colors = ['r', 'g', 'b', 'y']
unique_clusters = np.unique(clusters)

# Visualizar los resultados con leyenda
plt.figure(figsize=(10, 6))

# Colorear los puntos según el cluster asignado y mostrar en la leyenda
for cluster, color in zip(unique_clusters, colors):
    plt.scatter(X_pca[clusters == cluster, 0], X_pca[clusters == cluster, 1], c=color, label=f'Cluster {cluster}', edgecolor='k')

# Mostrar leyenda
plt.legend()

# Etiquetar los ejes
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Clustering de datos después de PCA con leyenda')

plt.show()
```
